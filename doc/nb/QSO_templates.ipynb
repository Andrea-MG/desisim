## Generate 100 

```python
seed = 123
nmodel = 10
```

```python
qso = QSO(minwave=3000)
```

```python
flux, wave, meta = qso.make_templates(nmodel=nmodel, zrange=(2.0, 4.0), seed=seed, N_perz=100,
                                      nocolorcuts=True, lyaforest=False)
```

```python
flux_forest, _, meta_forest = qso.make_templates(nmodel=nmodel, zrange=(2.0, 4.0), seed=seed, N_perz=100,
                                                 nocolorcuts=True, lyaforest=True)
```
